```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler,LabelEncoder
```

```python
numerical_features = adult.select_dtypes(include=['float64', 'int64']).columns

# Standard Scaling
standard_scaler = StandardScaler()
adult_standard_scaled = adult.copy()
adult_standard_scaled[numerical_features] = standard_scaler.fit_transform(adult[numerical_features])

# Display the first few rows of the Standard Scaled dataframe
print("\nStandard Scaled DataFrame:")
print(adult_standard_scaled.head())
```

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
adult_min_max_scaled = adult.copy()
adult_min_max_scaled[numerical_features] = min_max_scaler.fit_transform(adult[numerical_features])

# Display the first few rows of the Min-Max Scaled dataframe
print("\nMin-Max Scaled DataFrame:")
print(adult_min_max_scaled.head())
```

```python
categorical_columns = adult.select_dtypes(include=['object']).columns
# Apply One-Hot Encoding to categorical variables with less than 5 unique categories
for col in categorical_columns:
    if adult[col].nunique() < 5:
        adult= pd.get_dummies(adult, columns=[col], prefix=[col], drop_first=True)
# Display the first few rows of the dataframe after One-Hot Encoding
print("\nDataFrame after One-Hot Encoding:")
print(adult.head())
```

```python
categories_to_encode = adult.select_dtypes(include=['object']).columns  # Select categorical columns
for col in categories_to_encode:
    if adult[col].nunique() >5:# Check number of unique categories
        label_encoder = LabelEncoder()
        adult[col] = label_encoder.fit_transform(adult[col])

# Display the DataFrame after Label Encoding
print("\nDataFrame after Label Encoding:")
print(adult)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(adult['fnlwgt'], bins=10, edgecolor='pink')
plt.title('Histogram of Skewed Feature')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
# Apply log transformation
adult['log_feature1'] = np.log(adult['fnlwgt'])
adult['log_feature1']
```

```python
# Plot histogram of log-transformed feature
plt.subplot(1, 2, 2)
plt.hist(adult['log_feature1'], bins=10, edgecolor='pink')
plt.title('Histogram of Log-Transformed Feature')
plt.xlabel('Log Values')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Feature Selection**

```python
from sklearn.ensemble import IsolationForest
# Identify outliers using Isolation Forest
clf = IsolationForest(contamination=0.05, random_state=42)
numerical_features = adult.select_dtypes(include=['number'])
outlier_preds = clf.fit_predict(numerical_features.dropna())

outlier_mask = outlier_preds == -1
# Remove outliers from the dataframe
data_no_outliers = adult.loc[~outlier_mask]
# Display the dataframe after removing outliers
print("\nDataFrame after removing outliers:")
data_no_outliers
```
