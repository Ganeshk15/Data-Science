# **Standardization of Numerical Variables**

**Standardization :** Standardization, also known as Z-score normalization, is a technique used to scale features in a dataset so that they have the properties of a standard normal distribution. Specifically, the standardized data will have:
A mean (average) of 0
A standard deviation of 1
This process ensures that the data is on the same scale, which is important for many machine learning algorithms that are sensitive to the relative magnitude of the features, such as support vector machines (SVM), k-nearest neighbors (k-NN), and principal component analysis (PCA).
‚Äã

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
numerical_columns
```

```python
mean = sales[numerical_columns].mean()
std_dev = sales[numerical_columns].std()

# Apply the standardization formula
sales_standardized = (sales[numerical_columns] - mean) / std_dev

# Display the standardized data
print(sales_standardized)
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(sales[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column} (Before Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(sales_standardized[column], bins=20, color='salmon', edgecolor='black')
    plt.title(f'Distribution of {column} (After Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

# **Conversion of Categorical Data into Dummy Variables**

# ‚óè	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
In many machine learning algorithms, categorical variables cannot be directly used as inputs. One-hot encoding is a technique used to convert categorical variables into a format that can be provided to machine learning algorithms.
It creates binary (0 or 1) columns for each category in the categorical variable. This ensures that the categorical data is represented in a numerical format, allowing machine learning models to interpret and process it effectively.

```python
sales.categorical_columns=['City', 'SKU', 'BU', 'Brand', 'Model', 'Day']
print(sales.categorical_columns)
```

```python
sales = pd.get_dummies(sales.categorical_columns)
print(sales.head())
```
