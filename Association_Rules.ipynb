**1. What is lift and why is it important in Association rules?**

Lift is a key metric used in association rule learning, a popular data mining technique, particularly in market basket analysis. Lift measures the strength of an association rule over the expected likelihood of the rule, assuming the items are independent. It helps in determining how much more likely the occurrence of one item is given another, compared to if they were independent.

**Why is it important?**
 1. Identifying Strong Associations: Lift helps us identify relationships that are genuinely strong and not simply due to the popularity of individual items. A high lift value indicates a strong association between items, suggesting that purchasing one item significantly increases the likelihood of purchasing the other.

 2. Prioritizing Rules: In a large dataset with numerous association rules, lift helps prioritize the most interesting and valuable rules. We can focus on rules with higher lift values, as they represent the most significant and actionable insights.

 3. Improving Business Decisions: By understanding the lift between products, businesses can make better decisions regarding product placement, promotions, and targeted marketing campaigns. For example, if two products have a high lift value, a business might consider placing them near each other in a store or offering them as a bundle.


What is support and Confidence. How do you calculate them?

Support:
* Support is the fraction of transactions in the dataset that contain a specific itemset.
* It indicates the frequency of occurrence of an itemset.
* Formula: Support(X) = (Number of transactions containing X) / (Total number of transactions)
* Example: If an itemset "milk and bread" appears in 10 out of 100 transactions, its support is 0.1.

Confidence:
* Confidence measures the probability of another item being purchased given that a specific item has already been purchased.
* It indicates the conditional probability of an association rule.
* Formula: Confidence(X -> Y) = Support(X U Y) / Support(X)
* the confidence of the rule "milk -> bread" is 10/20 = 0.5.
